```python
from google.cloud import storage
client = storage.Client(project='696445312849')
```

```python
# Replace these values with your specific details
bucket_name = 'idd_dataset'
file_name = 'idd-detection.tar.gz'
destination_path = '/idd_dataset'
```

```python
from tqdm import tqdm
bucket = client.bucket(bucket_name)
blob = bucket.blob(file_name)
total_size = blob.size  # Get the total file size

with tqdm(total=total_size, unit='B', unit_scale=True, unit_divisor=1024) as pbar:
  with open(destination_path, 'wb') as f:
    for chunk in blob.download_as_string():
      if chunk is None:
        break
      f.write(chunk)
      pbar.update(len(chunk))

print(f"File downloaded to: {destination_path}")
```
